```python
import pandas as pd
from IPython.display import display

DATA_DIR = "/Users/alexandermittet/Library/Mobile Documents/com~apple~CloudDocs/uni_life/UniPi DAD/data_analytics_4_health_unipi/Data"

DATASETS = {
    "heart_diagnoses_1": "heart_diagnoses_1.csv",
    "laboratory_events_codes_2": "laboratory_events_codes_2.csv",
    "microbiology_events_codes_3": "microbiology_events_codes_3.csv",
    "procedure_code_4": "procedure_code_4.csv",
}

dataframes = {
    name: pd.read_csv(f"{DATA_DIR}/{filename}")
    for name, filename in DATASETS.items()
}
```

```python
for name, df in dataframes.items():
    print(f"===== {name} =====")
    display(df.head())
```

```python
FALSE_NA_MARKERS = {"none", "nan", "n/a", "null", "na", "n.a.", "-", ""}

for name, df in dataframes.items():
    print(f"===== {name} =====")
    obj_cols = df.select_dtypes(include="object").columns
    if not len(obj_cols):
        print("no object columns")
        continue

    for col in obj_cols:
        series = df[col].dropna()
        if series.empty:
            continue

        trimmed = series.astype(str).str.strip()
        coerced = pd.to_numeric(trimmed, errors="coerce")
        numeric_ratio = coerced.notna().mean()

        if numeric_ratio == 0:
            continue

        if numeric_ratio >= 0.5:
            print(f"column: {col} | numeric-like fraction: {numeric_ratio:.2f}")

            bad_mask = coerced.isna()
            if bad_mask.any():
                bad_tokens = trimmed[bad_mask].value_counts().head(10)
                false_na = [tok for tok in bad_tokens.index if tok.lower() in FALSE_NA_MARKERS]

                print("  problematic tokens:")
                for token, count in bad_tokens.items():
                    marker = " (false NA)" if token.lower() in FALSE_NA_MARKERS else ""
                    print(f"    {token!r}: {count}{marker}")

                if not false_na:
                    print("    none detected")
            else:
                print("  no problematic tokens")
```

```python
df2 = dataframes["laboratory_events_codes_2"].drop(columns=["value"], errors="ignore")
dataframes["laboratory_events_codes_2"] = df2
```

```python
def check_id_column(series):
    as_str = series.astype(str)
    trimmed = as_str.str.strip()
    whitespace_mask = as_str != trimmed
    decimal_mask = trimmed.str.contains(r"\.0$", na=False)
    return whitespace_mask, decimal_mask, trimmed

for name, df in dataframes.items():
    print(f"===== {name} =====")
    id_cols = [col for col in df.columns if "id" in col.lower()]
    if not id_cols:
        print("no id columns")
        continue

    for col in id_cols:
        whitespace_mask, decimal_mask, trimmed = check_id_column(df[col])
        whitespace_count = whitespace_mask.sum()
        decimal_count = decimal_mask.sum()

        print(f"column: {col}")
        if whitespace_count:
            print(f"  whitespace issues: {whitespace_count}")
            print(df.loc[whitespace_mask, col].head().to_list())
        else:
            print("  no whitespace issues")

        if decimal_count:
            print(f"  decimal .0 issues: {decimal_count}")
            print(trimmed.loc[decimal_mask].head().to_list())
        else:
            print("  no decimal .0 issues")
```

```python
DATETIME_KEYWORDS = ("time", "date", "timestamp", "datetime", "dod")

for name, df in dataframes.items():
    print(f"===== {name} =====")
    candidate_cols = [
        col for col in df.columns
        if df[col].dtype == "object" and any(key in col.lower() for key in DATETIME_KEYWORDS)
    ]

    if not candidate_cols:
        print("no obvious datetime columns")
        continue

    for col in candidate_cols:
        series = df[col]
        non_null = series.dropna()
        non_null_ratio = len(non_null) / len(series) if len(series) else 0
        if non_null.empty:
            print(f"column: {col} | all values missing")
            continue

        parsed = pd.to_datetime(non_null, errors="coerce", infer_datetime_format=True)
        success_ratio = parsed.notna().mean()
        print(
            f"column: {col} | parsed(non-null): {success_ratio:.2%} | availability: {non_null_ratio:.2%}"
        )
        if success_ratio < 0.9:
            sample_failed = non_null[parsed.isna()].astype(str).head(5)
            if not sample_failed.empty:
                print("  sample unparsed values:")
                for value in sample_failed:
                    print(f"    {value!r}")
        else:
            print("  parsed cleanly")
```

```python
import numpy as np

for name, df in dataframes.items():
    if "qc_flag" not in df.columns:
        continue

    qc_upper = df["qc_flag"].astype(str).str.upper()
    mask_fail = qc_upper == "FAIL"
    mask_ok = qc_upper == "OK"

    fail_count = mask_fail.sum()
    ok_count = mask_ok.sum()

    if fail_count == 0:
        print(f"{name}: no FAIL rows to compare")
        continue

    if ok_count == 0:
        print(f"{name}: no OK rows to compare")
        continue

    df_fail = df.loc[mask_fail]
    df_ok = df.loc[mask_ok]

    print(f"===== {name} =====")
    print(f"QC_FAIL rows: {fail_count} | QC_OK rows: {ok_count}")

    numeric_cols = df.select_dtypes(include=[np.number]).columns
    if len(numeric_cols) == 0:
        print("no numeric columns for outlier analysis")
    else:
        print("\nNaN comparison (numeric columns):")
        fail_nans = df_fail[numeric_cols].isna().sum().sum()
        ok_nans = df_ok[numeric_cols].isna().sum().sum()
        fail_nan_pct = (fail_nans / (len(df_fail) * len(numeric_cols))) * 100
        ok_nan_pct = (ok_nans / (len(df_ok) * len(numeric_cols))) * 100
        print(f"  QC_FAIL: {fail_nans} ({fail_nan_pct:.2f}%)")
        print(f"  QC_OK: {ok_nans} ({ok_nan_pct:.2f}%)")
        print(f"  Difference: {fail_nan_pct - ok_nan_pct:.2f}pp")

    print("\nOutlier comparison (z-score > 3):")
    outlier_counts_fail = {}
    outlier_counts_ok = {}

    for col in numeric_cols:
        fail_vals = df_fail[col].dropna()
        ok_vals = df_ok[col].dropna()

        if len(fail_vals) < 2 and len(ok_vals) < 2:
            continue

        if len(fail_vals) >= 2:
            fail_std = fail_vals.std()
            if fail_std > 0:
                z_fail = ((fail_vals - fail_vals.mean()) / fail_std).abs()
                outliers_fail = (z_fail > 3).sum()
                if outliers_fail > 0:
                    outlier_counts_fail[col] = outliers_fail
        if len(ok_vals) >= 2:
            ok_std = ok_vals.std()
            if ok_std > 0:
                z_ok = ((ok_vals - ok_vals.mean()) / ok_std).abs()
                outliers_ok = (z_ok > 3).sum()
                if outliers_ok > 0:
                    outlier_counts_ok[col] = outliers_ok

    if outlier_counts_fail or outlier_counts_ok:
        print("  Columns with outliers:")
        for col in sorted(set(list(outlier_counts_fail.keys()) + list(outlier_counts_ok.keys()))):
            fail_out = outlier_counts_fail.get(col, 0)
            ok_out = outlier_counts_ok.get(col, 0)
            fail_den = len(df_fail[col].dropna())
            ok_den = len(df_ok[col].dropna())
            fail_pct = (fail_out / fail_den) * 100 if fail_den else 0
            ok_pct = (ok_out / ok_den) * 100 if ok_den else 0
            print(f"    {col}: FAIL={fail_out} ({fail_pct:.1f}%) | OK={ok_out} ({ok_pct:.1f}%)")
    else:
        print("  no outliers detected")
    print()
```

```python
import math
import matplotlib.pyplot as plt

for name, df in dataframes.items():
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    if len(numeric_cols) == 0:
        print(f"{name}: no numeric columns to plot")
        continue

    cols_per_row = 3
    rows = math.ceil(len(numeric_cols) / cols_per_row)
    fig, axes = plt.subplots(rows, cols_per_row, figsize=(cols_per_row * 4, rows * 3))
    axes = axes.flatten()

    for ax, col in zip(axes, numeric_cols):
        data = df[col].dropna()
        if data.empty:
            ax.set_title(f"{col} (empty)")
            ax.axis("off")
            continue
        ax.boxplot(data, vert=True, patch_artist=True)
        ax.set_title(col)

    for idx in range(len(numeric_cols), len(axes)):
        axes[idx].axis("off")

    fig.suptitle(f"{name} - numeric columns", fontsize=14)
    fig.tight_layout()
    plt.show()
```

```python
from pandas.plotting import scatter_matrix

MAX_POINTS = 1000

for name, df in dataframes.items():
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    if len(numeric_cols) < 2:
        print(f"{name}: not enough numeric columns for scatter matrix")
        continue

    sample_df = df[numeric_cols].dropna()
    if len(sample_df) > MAX_POINTS:
        sample_df = sample_df.sample(MAX_POINTS, random_state=42)

    if sample_df.empty:
        print(f"{name}: numeric columns empty after dropping NaNs")
        continue

    axarr = scatter_matrix(sample_df, figsize=(12, 12), diagonal='hist')
    plt.suptitle(f"{name} - scatter matrix", y=0.92)
    for ax in axarr.flatten():
        ax.tick_params(labelsize=8)
    plt.show()
```
